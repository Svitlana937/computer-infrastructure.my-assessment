# Problem 1: Data from yfinance

````python
# Import necessary libraries.
# https://ranaroussi.github.io/yfinance/#quick-start
import yfinance as yf

# Import datetime library.
import datetime as dt

# Import os library.
import os

import pandas as pd
import matplotlib.pyplot as plt
````

````python
def get_data():
    """
    Downloads the last 5 days of hourly data for FAANG stocks
    and saves it into the 'data' folder with a timestamped filename.
    """
    # List of tickers we need.
    tickers = ["META", "AAPL", "AMZN", "NFLX", "GOOG"]

    # Get the current time for the filename.
    now = dt.datetime.now()

    df = yf.download(tickers, period="5d", interval="1h")
    

    # Build file name with CSV extension.
    path = "data/"
    file_name = now.strftime("%Y%m%d-%H%M%S") + ".csv"
    full_file_name = path + file_name

    

    # Save to CSV in the data folder. 
    df.to_csv(full_file_name)
````

````python
get_data()
````

## Problem 2: Plotting Data

````python

def plot_data():
    """
    Reads the latest CSV from 'data', plots Close prices of all stocks,
    and saves the plot in 'plots' folder with timestamp.
    """

    # Get all CSV files in the data folder
    files = [f for f in os.listdir('data') if f.endswith('.csv')]
    files.sort()  # sort alphabetically, last one is latest
    latest_file = files[-1]

    # Load CSV into DataFrame
    df = pd.read_csv(os.path.join('data', latest_file), header=[0,1], index_col=0, parse_dates=True)

    # Create a figure
    plt.figure(figsize=(10,6))

    # Plot Close prices (explicitly for clarity)
    plt.plot(df.index, df["Close"]["META"], label="META")
    plt.plot(df.index, df["Close"]["AAPL"], label="AAPL")
    plt.plot(df.index, df["Close"]["AMZN"], label="AMZN")
    plt.plot(df.index, df["Close"]["NFLX"], label="NFLX")
    plt.plot(df.index, df["Close"]["GOOG"], label="GOOG")

    # Labels, title, legend
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.title(f"Stock Prices as of {df.index[-1].date()}")
    plt.legend()

    # Plotting.
    timestamp = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
    plt.savefig("plots/" + timestamp + ".png")
    plt.close()
    
````

````python
plot_data()
````

## Problem 3: Script

### FAANG Stock Data Plotting Script

## Script Overview

The script `faang.py` does the following:

1. Finds the latest CSV file in the `data` folder.
2. Reads the Close prices for five FAANG stocks.
3. Plots all stock prices on a single figure.
4. Adds axis labels, a legend, and the date as the title.
5. Saves the plot as a PNG file in the `plots` folder with a timestamped filename.

## Problem 4: Automation with GitHub Actions 

To automate the plotting process, GitHub Actions is used.  
The workflow runs the `faang.py` script automatically and saves the generated plot to the repository.

### Workflow File Location

The workflow file is stored in:
`.github/workflows/github-actions-demo.yml`

### Workflow Triggers

The workflow is configured to run in two ways:

- **Manual trigger** using `workflow_dispatch`.
- **Scheduled trigger** using `cron`, which runs the workflow every Saturday at 08:00 UTC.

### Workflow Steps

The GitHub Actions workflow executes the following steps sequentially:

1. **Checkout repository**  
   Clones the repository so that the workflow has access to `faang.py` and other project files.

2. **Setup Python**  
   Installs Python 3.x on the GitHub runner.

3. **Install dependencies**  
   Installs all required Python packages listed in `requirements.txt`.

4. **Run FAANG script**  
   Executes `faang.py` to read the latest data file and generate the plot.

5. **Commit and push new plots**  
   Stages, commits, and pushes any new plot files to the repository automatically.

6. **List plots**  
   Optionally lists the contents of the `plots/` folder for verification.


### Permissions

The workflow uses write permissions:

```yaml
permissions:
  contents: write

## Automation Result

When the workflow runs:
- The latest CSV data file is read from the data/ folder.
- A new plot is created and saved in the plots/ folder.
- The plot is automatically committed and pushed to the repository's main branch.

This removes the need for manual execution.
